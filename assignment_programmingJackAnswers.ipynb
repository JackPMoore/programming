```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 50)
plt.plot(x)
plt.show()
```

```python
y = np.log(x)
z = np.exp(x)
plt.scatter(x, y)
plt.scatter(x, z)
plt.show()
```

```python
plt.xlabel("Exponential")
plt.ylabel("Natural Log")
plt.title("Natural Log and Exponential Functions")
plt.scatter(x, y, label = 'Natural Log')
plt.scatter(x, z, label = 'Exponential')
plt.legend(loc = 'lower right')
plt.show()
```

```python
x = np.linspace(-6.5, 6.5, 100)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x, y)
plt.scatter(x, z)
plt.show()
```

```python
x = np.linspace(-6.5, 6.5, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

```python
x = np.linspace(-6.5, 6.5, 100)
y = np.sin(x)
z = np.cos(x)
plt.xlabel("Cos")
plt.ylabel("Sin")
plt.title("Sin and Cos Functions")
plt.plot(x, y, label = 'Sine')
plt.plot(x, z, label = 'Cosine')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc = 'lower right')
plt.show()
```
